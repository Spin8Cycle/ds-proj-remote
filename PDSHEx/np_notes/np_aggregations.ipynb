# NumPy Aggregations

### Multidimensional Aggregates

```python
import numpy as np
```

```python
nprange = np.random.default_rng(seed=42)
```

```python
mda = nprange.integers(0, 10, (3, 4))
mda
```

```python
mda.sum()
```

```python
mda.max()
```

```python
mda.max(axis=1)
```

```python
mda.max(axis=0)
```

# NumPy Arrays Creation Basics

```python
import numpy as np
np.__version__
```

### A. Creating Arrays from Python Lists

```python
li = [2, 4, 6]
np.array(li)
```

```python
lf = [2.6, 4, 6]
np.array(lf)
```

```python
np.array(lf, dtype=int)
```

```python
np.array(li, dtype=np.float32)
```

```python
np.array([range(i, i+3) for i in [2, 4, 6]])
```

### B. Creating Arrays from Scratch

```python
np.zeros(10, dtype=int)
```

```python
np.ones((4, 7), dtype=float)
```

```python
np.full((6, 9), 69)
```

```python
np.arange(2, 28, 3) # (start, stop, step)
```

```python
np.linspace(0, 2.5, 5)
```

```python
np.random.random((4, 4)) # 4x4 array with random values between 0 and 1 
```

```python
np.random.normal(1, 3, (4,4)) # 4x4 array of normally distributed pseudorandom, with mean=1 and std=3
```

```python
np.random.randint(90, 100, (3,3)) # 3x3 array of pseudorandom integers in the interval [90, 100)
```

```python
np.eye(5)
```

```python
np.empty(4)
```

# Broadcasting

```python
import numpy as np
```

```python
a = np.arange(3)
b = np.arange(3)[:, np.newaxis]

print(a)
print(b)
```

```python
a + b
```

### A. Rules of Broadcasting

* Rule 1 :
    * If the two arrays differ in their number of dimensions, the shape of the one with fewer dimensions is padded with ones on its leading (left) side.

* Rule 2:
    * If the two arrays differ in their number of dimensions, the shape of the one with fewer dimensions is padded with ones on its leading (left) side.

* Rule 3:
    * If in any dimension the sizes disagree and neither is equal to 1, an error is raised.

##### A.1: Broadcasting Example 1

```python
M = np.ones((2, 3))
a = np.arange(3)
```

```python
print(M.shape)
print(M)
print(a.shape)
print(a)
```

```python
M + a
```

##### A.2: Broadcasting Example 2


```python
a = np.arange(3).reshape((3, 1))
b = np.arange(3)
```

```python
print(a.shape)
print(a)
print(b.shape)
print(b)
```

```python
a + b
```

##### A.3: Broadcasting Example 3

```python
M = np.ones((3, 2))
a = np.arange(3)
```

```python
print(M.shape)
print(M)
print(a.shape)
print(a)
```

```python
M+a
```

### B. Broadcasting in Practice

##### B.1 Centering an Array

```python
nprange = np.random.default_rng(seed=42)
x = nprange.random((10, 3))
x
```

```python
x_centered = x - x.mean(0)
x_centered
```

##### B.2 Plotting a 2-D Function

```python
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 50)[:, np.newaxis]
z = np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
plt.imshow(z, origin='lower', extent=[0, 5, 0, 5])
plt.colorbar()
```

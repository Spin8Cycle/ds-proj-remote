# Customizing Matplotlib: Configurations and Stylesheets

# A. Plot Customization by Hand:


```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('classic')
import numpy as np
```

```python
x = np.random.randn(1000)
plt.hist(x);
```

```python
# use a gray background
fig = plt.figure(facecolor='white')
ax = plt.axes(facecolor='#E6E6E6')
ax.set_axisbelow(True)

# draw solid white gridlines
plt.grid(color='w', linestyle='solid')

# hide axis spines
for spine in ax.spines.values():
    spine.set_visible(False)

# hide top and right ticks
ax.xaxis.tick_bottom()
ax.yaxis.tick_left()

# lighten ticks and labels
ax.tick_params(colors='gray', direction='out')
for tick in ax.get_xticklabels():
    tick.set_color('gray')

for tick in ax.get_yticklabels():
    tick.set_color('gray')
    
# control face and edge color of histogram
ax.hist(x, edgecolor='#E6E6E6', color='#EE6666');
```

# B. Changing the Defaults: `rcParams`

* Each time Matplotlib loads, it defines a runtime configuration containing the default styles for every plot element you create.
    * This configuration can be adjusted at any time using the  `plt.rc` convenience routine.

```python
from matplotlib import cycler

colors = cycler('color',
                ['#EE6666', '#3388BB', '#9988DD',
                '#EECC55', '#88BB44', '#FFBBBB'])

plt.rc('figure', facecolor='white')
plt.rc('axes', facecolor='#E6E6E6', edgecolor='none', axisbelow=True, grid=True, prop_cycle=colors)
plt.rc('grid', color='w', linestyle='solid')
plt.rc('xtick', direction='out', color='gray')
plt.rc('ytick', direction='out', color='gray')
plt.rc('patch', edgecolor='#E6E6E6')
plt.rc('lines', linewidth=2)
```

```python
plt.hist(x);
```

```python
for i in range(4):
    plt.plot(np.random.rand(10))
```

# C. Stylesheets

* A newer mechanism for adjusting overall chart styles is via Matplotlib’s style module, which includes a number of default stylesheets, as well as the ability to create and package your own styles. 
    * These stylesheets are formatted similarly to the .matplotlibrc files mentioned earlier, but must be named with a .mplstyle extension.

* To switch to a style sheet:
    * `plt.style.use('stylename')`

```python
plt.style.available[:5]
```

```python
def hist_and_lines():
    np.random.seed(0)
    fig, ax = plt.subplots(1, 2, figsize=(11, 4))
    ax[0].hist(np.random.randn(1000))
    for i in range(3):
        ax[1].plot(np.random.rand(10))
    ax[1].legend(['a', 'b', 'c'], loc='lower left')
```

```python
# Default Style
with plt.style.context('default'):
    hist_and_lines()
```

```python
# FiveThiryEight Style

with plt.style.context('fivethirtyeight'):
    hist_and_lines()
```

```python
# ggplot Style

with plt.style.context('ggplot'):
    hist_and_lines()
```

```python
# Bayesian Methods for hackers style

with plt.style.context('bmh'):
    hist_and_lines()
```

```python
# Dark Background Style

with plt.style.context('dark_background'):
    hist_and_lines()
```

```python
# Grayscale Style

with plt.style.context('grayscale'):
    hist_and_lines()
```

```python
# Seaborn Style

with plt.style.context('seaborn-whitegrid'):
    hist_and_lines()
```
